**Manejo de valores nulo:** Nuestra estrategia será reemplazar los valores nulos por la media o la mediana. Con la función df.describe() podemos ver los datos estadísticos de las columnas. En caso que la variable presente una distribución normal de los datos reemplazaremos por la media, caso contrario, reemplazaremos por la mediana.

```python
#Identificación de la cantidad de valores nulos que tiene cada columna en el data set actualizado

df.isnull().sum()
```

```python
# Reemplazar los valores nulos por la mediana en las columnas video_views_rank, country_rank, channel_type_rank, video_views_for_the_last_30_days	y created_year
df['video_views_rank'] = df['video_views_rank'].fillna(df['video_views_rank'].median())
df['country_rank'] = df['country_rank'].fillna(df['country_rank'].median())
df['channel_type_rank'] = df['channel_type_rank'].fillna(df['channel_type_rank'].median())
df['video_views_for_the_last_30_days'] = df['video_views_for_the_last_30_days'].fillna(df['video_views_for_the_last_30_days'].median())
df['created_year'] = df['created_year'].fillna(df['created_year'].median())
```

```python
# Reemplazar los valores nulos por la moda en las columnas 'Category', 'Channel_type' y 'Country'
df['category'] = df['category'].fillna(df['category'].mode()[0])
df['channel_type'] = df['channel_type'].fillna(df['channel_type'].mode()[0])
df['Country'] = df['Country'].fillna(df['Country'].mode()[0])
```

```python
# Verificación de ausencia de valores nulos luego del reemplazo de los mismos
df.isnull().sum()
```

```python
#Identificar outliers de las columnas de tipo numerico
columns_to_check = ['subscribers','video views','uploads','video_views_rank',
                         'country_rank','channel_type_rank','video_views_for_the_last_30_days',
                         'lowest_monthly_earnings','highest_monthly_earnings',
                         'lowest_yearly_earnings','highest_yearly_earnings',
                         ]

# Iterar sobre las columnas y calcular outliers
for column in columns_to_check:
    Q1 = df[column].quantile(0.25)  # Primer cuartil
    Q3 = df[column].quantile(0.75)  # Tercer cuartil
    IQR = Q3 - Q1  # Rango intercuartílico
    outliers = df[(df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))]  # Identificar outliers
    print(f"Outliers en la columna '{column}':")
    print(outliers)
    print("\n")
```

**Análisis:** Identificamos outliers en los valores numéricos, ya que estos valores extremos, originados por eventos excepcionales como canales virales, pueden distorsionar los patrones del dataset, afectar cálculos estadísticos y limitar la capacidad del modelo para generalizar.

```python
# Eliminar outliers
# Lista de columnas a analizar
columns_to_check = ['subscribers','video views','uploads','video_views_rank',
                         'country_rank','channel_type_rank','video_views_for_the_last_30_days',
                         'lowest_monthly_earnings','highest_monthly_earnings',
                         'lowest_yearly_earnings','highest_yearly_earnings',
                         ]


for column in columns_to_check:
    Q1 = df[column].quantile(0.25)  # Primer cuartil
    Q3 = df[column].quantile(0.75)  # Tercer cuartil
    IQR = Q3 - Q1  # Rango intercuartílico

    # Límites para valores no considerados outliers
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Filtrar el DataFrame para eliminar outliers
    df = df[(df[column] >= lower_limit) & (df[column] <= upper_limit)]

# Mostrar el DataFrame después de eliminar outliers
print("DataFrame después de eliminar los valores outliers:")
df
```

**Análisis:** Eliminamos los valores outliers ya que pueden confundir el análisis y hacer que el modelo predictivo sea menos preciso. Al eliminarlos, nos aseguramos de trabajar con datos más claros y útiles para obtener predicciones más confiables. La cantidad de datos se redujo a 528 filas.

En función del análisis que fuimos haciendo a lo largo del proyecto, seleccionamos las siguientes features para utilizar como predictores en nuestro modelo.

**Subscribers**: El número de suscriptores es un indicador clave del alcance del canal y, generalmente, se correlaciona con los ingresos.

**Video views**: Representa el interés del público en el contenido y tiene una relación directa con las ganancias por publicidad.

**video_views_for_the_last_30_days:** Refleja el rendimiento reciente del canal, relevante para predecir ingresos actuales.

**Category:** Diferentes tipos de contenido tienen diferentes tasas de monetización. Por ejemplo, categorías como tecnología o finanzas tienden a generar más ingresos que entretenimiento general.

**Country:** La ubicación del YouTuber influye en los ingresos debido a las diferencias en el CPM (Coste por Mil Impresiones) entre países.

**created_year:** El año de creación puede ser un proxy de la experiencia del canal en la plataforma, lo que influye en las ganancias.

Seleccionamos estas features porque tienen una relación directa o indirecta con las ganancias de un canal de YouTube. ## Evaluación del modelo

```python
# Calcula el R cuadrado
r_squared = r2_score(y_test, predictions)
print(f'R cuadrado: {r_squared}')
```

**Análisis:** El coeficiente de determinación obtenido es cercano a 1, por lo que el modelo ajusta bien y sirve para explicar los datos.

```python
# Calcular MSE (error cuadratico medio)
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)

# Imprimir el resultado
print(f'Test RMSE: {rmse}')
print(f'Test MSE: {mse}')
```

**Análisis:** El error cuadrático medio nos da una idea del error de predicción. Calculando la raiz del mismo obtenemos que el modelo tiene un error de más menos $549.914 de ganancias anuales de un canal de Youtube.

```python
media_ganancias = y_test.mean()
print(f"Media de las ganancias anuales: {media_ganancias}")
```

```python
#Cálculo del error relativo
error_relativo = rmse / media_ganancias *100
print(f"Error relativo: {error_relativo}")
```

**Análisis:**

El modelo tiene un error cuadrático medio (RMSE) de 549.914, lo que indica una desviación típica de las predicciones de aproximadamente esa cantidad en las ganancias anuales de un canal de YouTube.
